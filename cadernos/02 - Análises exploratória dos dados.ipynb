```python
corr_positiva = pd.DataFrame(columns=["Coluna",'Colunas_Corr','Valor_Corr'])
corr_negativa = pd.DataFrame(columns=["Coluna",'Colunas_Corr','Valor_Corr'])
```

### **Criando um laço for para retornar as correlações positivas e negativas das colunas**

```python
# print('Correlações das colunas numéricas')

for i in colunas_numericas:
    p = colunas_numericas.corr()[i].sort_values(ascending=False).head(4)
    n = colunas_numericas.corr()[i].sort_values(ascending=True).head(3)
    
    nova_linha_p = pd.DataFrame({"Coluna":i ,"Colunas_Corr":p.index.values, "Valor_Corr":p.values})
    nova_linha_n = pd.DataFrame({"Coluna":i ,"Colunas_Corr":n.index.values, "Valor_Corr":n.values})
    
    corr_positiva = pd.concat([corr_positiva, nova_linha_p], axis=0, ignore_index=True)
    corr_negativa = pd.concat([corr_negativa, nova_linha_n], axis=0, ignore_index=True)
    
    #print("Coluna:",i)
    #print(p)
    #print(n)
    #print('')
    #print('')
    
corr_positiva = corr_positiva.loc[corr_positiva['Valor_Corr'] != 1.0]
```

```python
for i in colunas_nao_numericas:
    print(f'Coluna: {i} | Valores únicos: {colunas_nao_numericas[i].nunique()}')
    
    a = colunas_nao_numericas[i].value_counts()
    fig,ax = plt.subplots(ncols=2, figsize=(9,3), tight_layout=True)
    ax[0].bar(a.index, a.values, color='darkblue')
    ax[0].set_title(f'Coluna: {i} | Valores únicos: {colunas_nao_numericas[i].nunique()}',fontsize=11, fontweight='bold')
    ax[0].yaxis.set_visible(False)
    ax[0].spines[['top','left','right']].set_visible(False)
    ax[0].set_xticks(range(len(a.index)))
    
    if len(a.index) > 6:
        for ii in range(0,len(a.index)):
            ax[0].annotate(a.values[ii], (a.index[ii], a.values[ii]+a.values[ii]*0.02), ha='center', fontsize=8)
        ax[0].set_xticklabels(a.index, rotation=90)
    else:
        for ii in range(0,len(a.index)):
            ax[0].annotate(a.values[ii], (a.index[ii], a.values[ii]+a.values[ii]*0.02), ha='center', fontsize=12)
    ax[1].pie(a.values, labels=a.index, autopct='%1.1f%%', pctdistance=0.75)
    plt.show()
    print('------------------------------------------------------')
    
```
