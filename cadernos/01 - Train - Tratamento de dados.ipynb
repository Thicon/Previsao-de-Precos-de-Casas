# **Tratando valores nulos**

#### **Descricões das 5 colunas com mais valores nulos:**
- **PoolQC** - Qualidade da piscina
- **MiscFeature** - recurso diverso não coberto em outras categorias
- **Alley** - Tipo de beco de acesso
- **Fence** - Qualidade da cerca
- **FireplaceQu** - Qualidade da lareira

#### **Descrição de alguns valores que serão analisados**
- Po: Poor
- Fa: Fair
- TA: Typical/Average
- Gd: Good
- Ex: Excellent

```python
# Visualizando os valores únicos das colunas com mais valores nulos
cols = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu']
for i in cols:
    print('Coluna:',i)
    print(base[i].value_counts())
    print(' ')
    print(base.groupby(i)['SalePrice'].median())
    print(' ')
    print(' ')
```

Analisando alguns dos dados acima vemos situações como:
- PoolQC - A qualidade realmente influencia na mediana do preço de venda da casa.
    - Fa: 171k, Ex: 490k ------ Porém, há pouquissimos valores (apenas 7)
- FireplaceQu - A qualidade da lareira influencia na mediana do preço de venda da casa.
    - Po: 131k, Ex: 314k ------ Diferença significativa

A coluna de área da piscina possui 1453 valores em 0 e apenas 7 com valores acima. Ou seja, os valores nulos na coluna PoolQC poderiam ser trocados para "No".
Porém, optarei por apagar essa coluna

```python
# Visualizando valores da coluna "PoolArea"
base['PoolArea'].value_counts()
```

Pela quantidade de valores nulos, optei por apagar as seguintes colunas no código abaixo:

```python
base = base.drop(['Id','MiscFeature','Alley','PoolQC','PoolArea','Fence'],axis=1)
```

```python
# Visualizando novamente as colunas com mais valores nulos
base.isnull().sum().sort_values(ascending=False).head(15)
```

### **Analisando as colunas de fogueira (Fireplace)**

```python
# Agrupando as colunas de fogueira e qualidade da fogueira
fogueiras = base.columns[base.columns.str.contains('Fireplace')].values
fogueiras = base[fogueiras]
fogueiras.groupby('Fireplaces')['FireplaceQu'].value_counts(dropna=False)
```

Como podemos ver acima, as linhas com valores nulos na coluna "FireplacesQu" são devido a não terem fogueiras na casa, e não por erro de digitação. Portando, vou atribuir o valor "No Fireplaces" para essas linhas

```python
base['FireplaceQu'] = base['FireplaceQu'].fillna("No Fireplaces")
```

### **Lot frontage**

```python
# Visualizando colunas que possuem a string "Lot"
lots = base.columns[base.columns.str.contains('Lot')].values
lots = base[lots]
lots.loc[lots.LotFrontage.isnull()]
```

```python
# Visualizando os valores de LotFrontage
base.LotFrontage.value_counts()
```

```python
# Gráfico de caixa da coluna LotFrontage
dados_lotfrontage = base[['LotFrontage','SalePrice']].copy()
sns.boxplot(dados_lotfrontage[['LotFrontage']])
```

#### Dividindo os valores da coluna "LotFrontage" em grupos

```python
maximo = dados_lotfrontage['LotFrontage'].max()
minimo = dados_lotfrontage['LotFrontage'].min()
dados_lotfrontage['Faixas'] = pd.cut(dados_lotfrontage['LotFrontage'],
                                    bins=[minimo,60,100,140,180,maximo],
                                    labels=['20-60','60-100','100-140','140-180','180-313'],
                                    include_lowest=True, right=False)
```

```python
# Visualizando as medianas desses grupos
b = dados_lotfrontage.groupby('Faixas')['SalePrice'].median()
b
```

#### **Visualizando as relações entre "LotFrontage" e "SalePrice"**

```python
fig, ax = plt.subplots(figsize=(12,5), ncols=2)
ax[0].plot(b.index, b.values)
ax[0].set_title("X = Lot Frontage --- y = Sale Price --- Mediana")
ax[1].scatter(base['LotFrontage'], base['SalePrice'], s=7)
ax[1].set_title("Lot Frontage e Sale Price (Entre 20 e 200)")
ax[1].set_xlim(20,200)
plt.show()
```

Com bases na análise acima, vemos uma correlação entre o SalePrice e a coluna LotFrontage. Ou seja, será que realmente devo apagá-la?

```python
# Criando duas bases: Uma com valores nulos dessa coluna, e uma sem.
nulos = lots.loc[lots['LotFrontage'].isnull()].copy()
naonulos = lots.loc[lots['LotFrontage'].notnull()].copy()
nulos.head()
```

```python
# Visualizando a base de valores nulos em "LotFrontage"
naonulos.head()
```

```python
# Aplicando os grupos na base de valores não-nulos
maximo = base['LotFrontage'].max()
minimo = base['LotFrontage'].min()
naonulos['Faixas'] = pd.cut(naonulos['LotFrontage'],
                                    bins=[minimo,60,100,140,180,maximo],
                                    labels=['20-59','60-99','100-139','140-179','180-313'],
                                    include_lowest=True, right=False)
naonulos.head()
```

```python
# Agrupando os valores de colunas "Lot", visualizando as medianas
naonulos.groupby(['LotConfig','LotShape'])['LotFrontage'].median()
```

Aplicando as medianas nos valores nulos da coluna "LotFrontage", de acordo com os valores dos agrupamentos

```python
base['LotFrontage'] = base.LotFrontage.fillna(base.groupby(['LotConfig','LotShape'])['LotFrontage'].transform('median'))
```
